```python
# Первый шаг: Импортируем необходимые библиотеки
import pandas as pd
from sklearn.linear_model import LinearRegression
import plotly.graph_objects as go
from plotly.subplots import make_subplots


df = pd.read_csv('Salary_dataset.csv')
print(df.describe())
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

model = LinearRegression()
model.fit(df[['YearsExperience']], df['Salary'])
plt.scatter(df['YearsExperience'], df['Salary'], color='blue')
plt.plot(df['YearsExperience'], model.predict(df[['YearsExperience']]), color='red')
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
hf = pd.read_csv('heart.csv')
ss = pd.read_csv('supermarket_sales.csv')

# Plotting linear regression for the heart.csv dataset (Age vs. MaxHR)
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='MaxHR', data=hf)
plt.title('Linear Regression: Age vs MaxHR')
plt.xlabel('Age')
plt.ylabel('MaxHR (Maximum Heart Rate)')
plt.show()

# Plotting linear regression for the supermarket_sales.csv dataset (Unit Price vs. Quantity)
plt.figure(figsize=(10, 6))
sns.regplot(x='Unit price', y='Quantity', data=ss)
plt.title('Linear Regression: Unit Price vs Quantity')
plt.xlabel('Unit Price')
plt.ylabel('Quantity')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
hf = pd.read_csv('heart.csv')
ss = pd.read_csv('supermarket_sales.csv')

# Plotting linear regression for the heart.csv dataset (Cholesterol vs. RestingBP)
plt.figure(figsize=(10, 6))
sns.regplot(x='Cholesterol', y='RestingBP', data=hf)
plt.title('Linear Regression: Cholesterol vs RestingBP')
plt.xlabel('Cholesterol')
plt.ylabel('RestingBP (Resting Blood Pressure)')
plt.show()

# Plotting linear regression for the supermarket_sales.csv dataset (Total vs. Rating)
plt.figure(figsize=(10, 6))
sns.regplot(x='Total', y='Rating', data=ss)
plt.title('Linear Regression: Total vs Rating')
plt.xlabel('Total Sales')
plt.ylabel('Rating')
plt.show()
```
